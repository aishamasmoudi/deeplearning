```python
from googletrans import Translator
import pandas as pd
import time
```

```python
def translate_text(text, src, dest):
    translator = Translator()
    time.sleep(0.3)
    return translator.translate(text, src=src, dest=dest).text
```

## Translation

Load the file containing the listings you want to translate

```python
data_ar = pd.read_csv("/content/ar_dataset_600.csv")
```

```python
data_ar_en = []
```

```python
for i in data_ar.index:
  try:
    tweet = translate_text(data_ar['tweet'][i], "ar", "en")
    data_ar_en.append([i, tweet])
  except:
    print("There was a mistake at index:", i)
```

```python
data_ar
```

```python
data_ar_en = pd.DataFrame(data_ar_en, columns=['index', 'tweet'])
data_ar['index'] = data_ar.index
data_ar_en_merge = data_ar.merge(data_ar_en, on="index")
data_ar_en_merge
```

```python
data_ar_en_merge['tweet'] = data_ar_en_merge['tweet_y']
data_ar_en_merge = data_ar_en_merge.drop(['tweet_x', 'tweet_y', 'index'], axis =1)
data_ar_en_merge
```

```python
data_ar_en_merge.to_csv("ar_dataset_600_translated.csv")
```
